## Case 2

```python


radius = 1 

center = (0, 0)  
num_points_sparse = 20  
rotation_angle   = np.pi 
theta_sparse = np.linspace(0, 2 * np.pi, num_points_sparse)

x_1 = center[0] + radius * np.cos(theta_sparse)
y_1 = center[1] + radius * np.sin(theta_sparse)

radius  = radius + 0.2

x_2 = center[0] + radius * np.cos(theta_sparse + rotation_angle)
y_2 = center[1] + radius * np.sin(theta_sparse + rotation_angle)

radius  = radius + 0.2
rotation_angle =+ 1
x_3 = center[0] + radius * np.cos(theta_sparse + rotation_angle)
y_3 = center[1] + radius * np.sin(theta_sparse + rotation_angle)

x_combined = np.concatenate((x_1, x_2 , x_3))
y_combined = np.concatenate((y_1, y_2 , y_3))

plt.figure(figsize=(10, 10))
plt.scatter(x_combined, y_combined)
plt.title('Circles with Different Radii')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()

```

## Case 3

```python
def possible_circle_divisions(num_points):
    divisions = []
    for i in range(1, num_points + 1):
        if num_points % i == 0:
            divisions.append(i)
    
    
    if divisions:
        average_division = sum(divisions) / len(divisions)
        
        return int(average_division)
    return  0  

```

```python

def gen_circles_points(radius: float, radius_increase: float, num_points: int, default_rotation_angle: float, center: Tuple[float, float]): 
    x = [] 
    y = []
    rotation_angle =  0
    num_circle = possible_circle_divisions(num_points)
    
    for circle in range(1, num_circle + 1):
        gen_angles = np.linspace(0, 2 * np.pi, int(num_points / num_circle))  
        
        random_rotations = np.random.uniform(0, default_rotation_angle , len(gen_angles))
        
        gen_x = center[0] + radius * np.cos(gen_angles + random_rotations)
        gen_y = center[1] + radius * np.sin(gen_angles + random_rotations)  
        
        x = np.concatenate((x, gen_x))
        y = np.concatenate((y, gen_y))
        
        radius += radius_increase

    sns.set(style="whitegrid")

    plt.figure(figsize=(20, 20))
    plt.scatter(x, y, c='#6256CA', s=100, alpha=0.7) 
    plt.title('Circles with Different Radii', fontsize=20)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()


gen_circles_points(radius=5.0, radius_increase=0.5, num_points=200, default_rotation_angle=2 * np.pi, center=(1.0, 6.0))
```
