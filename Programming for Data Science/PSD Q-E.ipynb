```python
import numpy as np
import matplotlib.pyplot as plt

# إنشاء مصفوفة 5 أبعاد
array_5d = np.random.randint(0, 100, size=(4, 3, 2, 5, 6))  # 4x3x2x5x6

# عرض شكل المصفوفة
array_shape = array_5d.shape

# إسقاط البعدين الرابع والخامس لتحويل البيانات إلى 2D لأغراض التصور
array_2d = array_5d[0, :, 0, :, :]

# عرض المصفوفة المسطحة عبر التصور
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(array_2d.reshape(array_2d.shape[0], -1), cmap="viridis", aspect="auto")
ax.set_title("Visualization of Flattened 5D Array to 2D")
plt.colorbar(im, ax=ax)
plt.show()

array_shape
```

```python
from mpl_toolkits.mplot3d import Axes3D

# إنشاء بيانات ثلاثية الأبعاد
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# إنشاء الرسم ثلاثي الأبعاد
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# تخصيص الرسم
ax.set_title("3D Surface Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
plt.show()
```

```python
# عرض مصفوفة ثلاثية الأبعاد باستخدام الرسوم ثلاثية الأبعاد
array_3d = np.random.randint(0, 100000, size=(100,100, 100))  # مصفوفة 5x5x5


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# الحصول على الإحداثيات والقيم من المصفوفة
x, y, z = np.indices(array_3d.shape)
values = array_3d.flatten()

# رسم النقاط
sc = ax.scatter(x.flatten(), y.flatten(), z.flatten(), c=values, cmap='viridis', s=100)

# تخصيص الرسم
ax.set_title("3D Visualization of the Array")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
fig.colorbar(sc, ax=ax, shrink=0.5, aspect=10)
plt.show()
```

### Join

```python
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
}, index=['a', 'b'])

df2 = pd.DataFrame({
    'Score': [85, 90]
}, index=['a', 'b'])

joined_df = df1.join(df2)
print(joined_df)
```

### Concat

```python
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
})

df2 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Age': [35, 40]
})

concatenated_df = pd.concat([df1, df2], ignore_index=True)
print(concatenated_df)
```

### Merge

- inner

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'] , 
    'NIN' : [213 ,547 ,57]
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [30, 35, 40]
})

merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

--------

- outer 

```python
merged_df = pd.merge(df1, df2, on='ID', how='outer')
print(merged_df)
```

--------

- Left

```python
merged_df = pd.merge(df1, df2, on='ID', how='left')
print(merged_df)
```

-----

- Rigth

```python
merged_df = pd.merge(df1, df2, on='ID', how='right')
print(merged_df)
```
