```python
def dimensions(image_path : str) :
    with open(image_path, 'rb') as f:
        f.seek(16) 
        width = int.from_bytes(f.read(4), 'big')  
        height = int.from_bytes(f.read(4), 'big')   
    
    return width, height
```

```python
print(dimensions(image_path))
```
