## Convert To Matrix

```python
def jpg_to_matrix(filename):
    with open(filename, 'rb') as f:
        data = f.read()

    matrix = []
    width = 0
    height = 0
    idx = 0

    # Extract width and height
    while idx < len(data):
        if data[idx] == 0xFF:
            marker = data[idx + 1]
            if marker == 0xC0 or marker == 0xC2:  # SOF marker (Start of Frame)
                height = (data[idx + 5] << 8) + data[idx + 6]
                width = (data[idx + 7] << 8) + data[idx + 8]
                idx += 9  # Move to the start of the image data
                break
            idx += 2
        else:
            idx += 1

    if width == 0 or height == 0:
        raise ValueError("Unable to determine image dimensions.")

    idx += 1  # Skip the marker

    # Read pixel data
    for h in range(height):
        row = []
        for w in range(width):
            if idx + 2 < len(data):  # Check if there's enough data
                pixel = (data[idx], data[idx + 1], data[idx + 2])  # RGB
                row.append(pixel)
                idx += 3
            else:
                break
        # Pad the row if it is shorter than the expected width
        while len(row) < width:
            row.append((0, 0, 0))  # Add black pixels (or any default value)
        matrix.append(row)

    return matrix, width, height
```

## Calculate the brightness of IMG

```python
def calculate_brightness(IMG ,img_height , img_width  ):
    R_values = []
    G_values = []
    B_values = []
    brightness_values = [] 
    for i in range(0 , img_height) : 
        for j in range(0 , img_width) : 
            R_values.append(IMG[i][j][0])
            G_values.append(IMG[i][j][1])
            B_values.append(IMG[i][j][2])
            brightness = (0.29 * IMG[i][j][0] + 0.58 * IMG[i][j][1] + 0.11 * IMG[i][j][2])
            brightness_values.append(brightness)
            
    average_brightness = sum(brightness_values) / len(brightness_values)
    
    return  average_brightness , brightness_values
    
    
```

```python
        
average_brightness ,brightness_values = calculate_brightness(IMG,img_height,img_width)

print(f'Average Brightness: {average_brightness}')
print(f' Brightness values : {brightness_values[:8]}')
```

## Calculate the contrast of IMG

```python
import math 

def calculate_contrast(average_brightness , brightness_values , N , M) : 
    
    contrast_sum = 0
    for brightness in brightness_values:
        
        contrast_sum += (brightness - average_brightness) ** 2
        
    contrast  =  math.sqrt(contrast_sum / (M * N))
    
    return contrast
    
```

```python
print(f'Contrast: {calculate_contrast(average_brightness , brightness_values , img_width , img_height )}')
```

## Show the each of the RGB channels separately

```python
def rgb_matrix(IMG , img_height , img_width  ) : 
    
    R_matrix = []
    G_matrix = []
    B_matrix = []
    
    for i in range(0 , img_height) :
        R_row = []
        G_row = []
        B_row = []
        for j in range(0 , img_width) :
            R_row.append(IMG[i][j][0])
            G_row.append(IMG[i][j][1])
            B_row.append(IMG[i][j][2])
        
        R_matrix.append(R_row)
        G_matrix.append(G_row)
        B_matrix.append(B_row)
    
    return R_matrix , G_matrix , B_matrix
```

```python
R_matrix , G_matrix , B_matrix = rgb_matrix(IMG , img_height , img_width  )
```
